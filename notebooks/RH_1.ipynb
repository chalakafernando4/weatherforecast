## Importing the libraries

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

from keras.models import Sequential
import matplotlib.patches as mpatches
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

import seaborn as sns
plt.style.use('fivethirtyeight')
```

## Data reading and preprocessing

```python
data=pd.read_csv('../dataset/Seethaeliya1.csv')
##importing the data
```

```python
data.head()
##watching head values of dataset
```

```python
data.shape
##the shape of the dataset
```

```python
##showing the datatypes
data.dtypes
```

```python
##converting date and time
data['Date']=pd.to_datetime(data['Date'])
```

```python
data.dtypes
```

```python
##We will need to remove the T_max feature because there are so much null values
##remove the T_Max column
data=data.drop(['T_Max'],axis=1)
```

```python
data.set_index('Date',inplace=True) ##indexing
```

```python
##time series plot of minimum rh1
plt.figure(figsize=(20,8))
plt.plot(data['RH_1'])
plt.title('Time Series graph of minimum Relative Humidity at 8.30 a.m')
plt.xlabel('Date')
plt.ylabel('Relative Humidity')
plt.show()
```

```python
data.head()
```

```python
##create a new dataframe with only the RH_1 column
data2=data.filter(['RH_1'])
data2=data2.interpolate() ##handle null values with linear interpolation
##convert the dataframe into a numpy array
dataset2=data2.values
##get the number of rows to train the model
train_data_len=math.ceil(len(dataset2)*.8)

train_data_len
```

```python
data2
```

```python
data2.apply(pd.isnull).sum() ##check null values after interpolation
```

```python
##time series plot of minimum rh1
plt.figure(figsize=(20,8))
plt.plot(data2['RH_1'])
plt.title('Time Series graph of minimum Relative Humidity at 8.30 a.m')
plt.xlabel('Date')
plt.ylabel('Relative Humidity')
plt.show()
```

```python
##Scaling the data
scaler=MinMaxScaler(feature_range=(0,1))
sc=scaler.fit_transform(dataset2)
sc
```

```python
#Create the training data set
##create the scaled training data set
train_data=sc[0:train_data_len,:]
##split the data into x_train and y_train data sets

x_train=[] ##features
y_train=[] ##target

for i in range(60,len(train_data)):
    x_train.append(train_data[i-60:i,0])
    y_train.append(train_data[i,0])
    if i<=60:
        print(x_train)
        print(y_train)
        print()
```

```python
##convert train sets to numpyarrays
x_train,y_train=np.array(x_train),np.array(y_train)
x_train
y_train
##here we have 2d array
##but dealing with LSTM we need to have 3d array
##therefore it is needed to reshape the data
```

```python
##reshape the data
x_train=np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
x_train.shape
```

## Model Building

```python
model=Sequential()
model.add(LSTM(50,activation='relu',return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(LSTM(50,activation='relu',return_sequences=False))
model.add(Dense(25,activation='relu'))
model.add(Dense(1))

#compile the model
model.compile(optimizer='adam',loss='mean_squared_error')

##train the model
history = model.fit(x_train,y_train,epochs=10,batch_size=1)
```

```python
plt.figure(figsize=(8,5))
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Model Accuracy")
```

```python
##create the testing dataset
##create a new array containing scaled values from index 1694 to 2192
test_data=sc[train_data_len-60:,:]
##Create data set x_test and y_test
x_test=[]
y_test=dataset2[train_data_len:,:]

for i in range(60,len(test_data)):
    x_test.append(test_data[i-60:i,0])
    
```

```python
##convert the data into numpy array
x_test=np.array(x_test)
```

```python
##reshape the dataset
x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
```

```python
##get the models predicted temperatures
predictions=model.predict(x_test)
predictions=scaler.inverse_transform(predictions) ##unscaling

```

```python
##calculate RMSE
##lower the RMSE higher the accuracy
rmse=np.sqrt(np.mean(((predictions-y_test)**2)))
rmse
```

```python
##plot the data
train=data2[:train_data_len]
validation_data=data2[train_data_len:]
validation_data['Predictions']=predictions

##visualise the data
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date')
plt.ylabel('Relative Humidity')
plt.plot(train['RH_1'])
plt.plot(validation_data[['RH_1','Predictions']])
plt.legend(['Train','Val','Predictions'],loc='lower right')
plt.show()
```
